```python
import torch
import torch.nn as nn

linear = nn.Linear(3, 2)  # 3 input features → 2 output features
x = torch.randn(1, 3)  # Batch size 1, 3 features
y = linear(x)
print(y.shape)  # Output: torch.Size([1, 2])
```

```python
activation = nn.Sigmoid()
x = torch.tensor([-1.0, 5.0, -0.5])
y = activation(x)
print(y)  # Output: tensor([0., 2., 0.])
```

```python
loss_fn = nn.MSELoss()
logits = torch.tensor([[2.0, 1.0, 0.1]])  # Logits (before Softmax)
target = torch.tensor([0])  # Class index
loss = loss_fn(logits, target)
print(loss.item())
```

```python
import torch
import torchvision.transforms as transforms
from torchvision import models
from PIL import Image

# Load a pre-trained model
model = models.resnet18(pretrained=True)
model.eval()  # Set to evaluation mode

# Load and preprocess an image
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

image = Image.open("/content/drive/MyDrive/DL/colorful-790254_640.jpg")
image = transform(image).unsqueeze(0)  # Add batch dimension

# Make prediction
with torch.no_grad():
    output = model(image)
    predicted_class = torch.argmax(output, dim=1)
print(predicted_class)
```
